```
class Maxout():
  def __init__(self, in_features, out_features, n_channels, bias=True):
    super().__init__()

    self.in_features = in_features
    self.out_features = out_features
    self.n_channels = n_channels

    self.weight =tf.Variable(n_channels * out_features, in_features)
    self.bias = tf.Variable(n_channels * out_features)

  def forward(self, input):
    a = tf.matmul(input, self.weight)+self.bias
    b = tf.nn.max_pool1d(a.unsqueeze(-3),self.n_channels)
    return b.squeeze()

```
